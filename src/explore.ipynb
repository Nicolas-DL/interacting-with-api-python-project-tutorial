```python
import os
import pandas as pd
import seaborn as sns
from dotenv import load_dotenv

# load the .env file variables
load_dotenv()

```

```python
client_id = os.environ.get('CLIENT_ID')
client_secret = os.environ.get('CLIENT_SECRET')
```

```python
print(client_id)
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

conn = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id, client_secret=client_secret))
```

```python
id_nirvana = '6olE6TJLqED3rqDCT0FyPh'

response = conn.artist_top_tracks(id_nirvana)
print(response['tracks'])
```

```python
df_tracks = pd.DataFrame.from_records(response['tracks'])
df_tracks
```

```python
scatter = sns.scatterplot(data=df_tracks, x='popularity', y='duration_ms')

fig = scatter.get_figure()
fig.savefig('scatter.png')
```
